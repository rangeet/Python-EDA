### 3.10) Aggregating (Groupby)

```python
Stores = pd.read_csv('E:/DataSets/stores.csv')
```

```python
Stores.columns
```

### Eg1 : Group by Location wrt TotalSales (sum)

```python
temp = Stores.groupby(by=['Location'])
result1 = temp[['TotalSales']].sum().add_prefix('SumOf_')   #pd.plot

result1 = temp[['TotalSales']].sum().add_prefix('SumOf_').reset_index()   #seaborn
# result1 = temp[['TotalSales']].sum().add_prefix('SumOf_').reset_index(drop=True)
result1
```

### Eg2 : Group by Location and StoreType wrt OpCost (mean)

```python
temp = Stores.groupby(by=['Location','StoreType'])
result2 = temp[['OperatingCost']].mean().add_prefix('MeanOf_')   
result2 = temp[['OperatingCost']].mean().add_prefix('MeanOf_').reset_index()
result2
```

### Eg3 : Group by Location wrt OpCost and TotalSales (mean)

```python
temp = Stores.groupby(by=['Location'])
result3 = temp[['OperatingCost','TotalSales']].mean().add_prefix('MeanOf_')   
result3 = temp[['OperatingCost','TotalSales']].mean().add_prefix('MeanOf_').reset_index()
result3
```

### Eg4 : Group by Location wrt OpCost (mean) and TotalSales (sum)

#### Method 1

```python
temp = Stores.groupby(by=['Location'])
a = temp[['OperatingCost']].mean().add_prefix('MeanOf_')   
b = temp[['TotalSales']].sum().add_prefix('SumOf_')
result4a = pd.concat(objs=[a,b],axis=1).reset_index()
result4a
```

#### Method 2

```python
temp = Stores.groupby(by=['Location'])
result4b = temp[['OperatingCost','TotalSales']].agg({'OperatingCost':'mean',
                                          'TotalSales':'sum'})

result4b = result4b.rename(columns={'OperatingCost':'MeanOf_OperatingCost','TotalSales':'SumOf_TotalSales'})
result4b = result4b.reset_index()
result4b
```
